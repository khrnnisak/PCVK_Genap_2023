```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pylab as plt
```

```python
urls = ["https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg",
        "https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg",
        "https://placekitten.com/800/571"]

for url in urls:
  image = io.imread(url)
  image = cv.resize(image, (0,0), fx=0.5, fy=0.5)
  image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
  final_frame = cv.hconcat((image, image_2))
  cv2_imshow(final_frame)
  print('\n')
```

```python
tinggi = image_2.shape[0]
lebar = image_2.shape[1]
print("resolusi image : tinggi x lebar = ", tinggi, " x ", lebar)
cv2_imshow(image_2)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

for y in range (lebar):
  image_3[int((tinggi)/2),y] = [255,255,255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

```python
image_4 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_5 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

for x in range (tinggi):
  image_5[x, int((lebar)/2)] = [255,255,255]

final_frame2 = cv.hconcat((image_2, image_5))
cv2_imshow(final_frame2)
```

```python
image_6 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_7 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

for x in range (tinggi):
  image_7[(x,x)] = [255,255,255]
  image_7[(x, tinggi-x-1)] = [255,255,255]
diag = cv.hconcat((image_6, image_7))
cv2_imshow(diag)
```

```python

image_8 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_9 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

y1 = int(lebar/4)
y2 = int(lebar/4*2)

for y in range (y1, y2):
    image_9[int(tinggi/2), y] = [255, 255, 255]

final_frame3 = cv.hconcat((image_8, image_9))
cv2_imshow(final_frame3)
```

```python
image_10 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_11 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

y1 = int(lebar/4)
y2 = int(lebar/4*2)
x1 = int(tinggi/4)
x2 = int(tinggi/4*2)

for x in range (x1, x2):
  for y in range(y1,y2):
    image_11[x, y] = [255, 255, 255]

final_frame4 = cv.hconcat((image_10, image_11))
cv2_imshow(final_frame4)
```
