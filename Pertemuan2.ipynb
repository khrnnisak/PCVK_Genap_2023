2. Apakah perbedaan dan pengaruhnya pembuatan black image antara tipe data int16 dan int32?
**Jawab** : Perbedaan terlihat dari ukuran filenya. Gambar dengan tipe data int16 membutuhkan 16 bit / 2 byte untuk menyimpan satu nilai, sedangkan tipe data int32 membutuhkan 32 bit / 4 byte. Gambar yang betipe data int32 akan memiliki ukuran file yang lebih besar dibandingkan dengan gambar bertipe data int16. Hal ini karena membutuhkan lebih banyak memori untuk menyimpan suatu nilai.
